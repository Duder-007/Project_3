```python
import pandas as pd

# Read the content of the biblical_feminine_names.csv file
with open('Baby_Names_Start/biblical_feminine_names.csv', 'r') as file:
    biblical_feminine_content = file.read()

# Read the content of the biblical_masculine_names.csv file
with open('Baby_Names_Start/biblical_masculine_names.csv', 'r') as file:
    biblical_masculine_content = file.read()
```

```python
# Display the first few lines of each file to verify their structure
print("Biblical Feminine Names Content:")
print('\n'.join(biblical_feminine_content.split('\n')[:10]))

print("Biblical Masculine Names Content:")
print('\n'.join(biblical_masculine_content.split('\n')[:10]))
```

```python
# Load the new CSV files
biblical_feminine_names = pd.read_csv('Baby_Names_Start/biblical_feminine_names.csv')
biblical_masculine_names = pd.read_csv('Baby_Names_Start/biblical_masculine_names.csv')
```

```python
# Combine the biblical names lists and ensure they are capitalized
biblical_names_list = pd.concat([biblical_feminine_names, biblical_masculine_names])['Name'].str.strip().str.capitalize().tolist()
```

```python
# Print the combined list of biblical names for debugging
print("Combined Biblical Names List:")
print(biblical_names_list)
```

```python
# Load the top baby names by state CSV file
top_baby_names_by_state = pd.read_csv('Baby_Names_Start/Top_baby_names_by_state.csv')
```

```python
# Ensure the names in the top_baby_names_by_state DataFrame are properly capitalized and stripped
top_baby_names_by_state['Name'] = top_baby_names_by_state['Name'].str.strip().str.capitalize()
```

```python
# Debug: Check a few rows of the dataset after capitalization
print("Top Baby Names by State (Capitalized) Head:")
print(top_baby_names_by_state.head(20))
```

```python
# Filter the top_baby_names_by_state for names that are in the biblical names list
filtered_names = top_baby_names_by_state[top_baby_names_by_state['Name'].isin(biblical_names_list)]
```

```python
# Debug: Check the filtered dataset
print("Filtered Names Head:")
print(filtered_names.head(20))
```

```python
# Check for names that should be in the filtered list but are missing
missing_names = [name for name in biblical_names_list if name not in filtered_names['Name'].unique()]
print("Missing Biblical Names in Filtered Data:")
print(missing_names)
```

```python
# Group by Name and Gender, and calculate the sum of counts
biblical_name_counts = filtered_names.groupby(['Name', 'Gender'])['Count'].sum().reset_index()
```

```python
# Debug: Check the biblical name counts
print("Biblical Name Counts:")
print(biblical_name_counts)
```

```python
# Calculate the total counts for each gender
total_counts = top_baby_names_by_state.groupby(['Gender'])['Count'].sum().reset_index()
total_counts = total_counts.rename(columns={'Count': 'Total_Count'})
```

```python
# Debug: Check the total counts by gender
print("Total Counts by Gender:")
print(total_counts)
```

```python
# Merge the biblical name counts with the total counts
merged_data = pd.merge(biblical_name_counts, total_counts, on='Gender')
```

```python
# Calculate the percentage
merged_data['Percentage'] = (merged_data['Count'] / merged_data['Total_Count'] * 100).round(2)
```

```python
# Display the results
print("Merged Data:")
print(merged_data)
```

```python
# Check the number of unique female and male biblical names with non-zero counts
unique_female_names = merged_data[merged_data['Gender'] == 'F']
unique_male_names = merged_data[merged_data['Gender'] == 'M']
print(f"Number of unique female biblical names with non-zero counts: {len(unique_female_names)}")
print(f"Number of unique male biblical names with non-zero counts: {len(unique_male_names)}")
```

```python
# Get the top 5 male and top 5 female percentages
top_5_biblical_male = merged_data[merged_data['Gender'] == 'M'].nlargest(5, 'Percentage')
top_5_biblical_female = merged_data[merged_data['Gender'] == 'F'].nlargest(5, 'Percentage')

```

```python
# Display the top 5 results
print("Top 5 Male Percentages:")
print(top_5_male)

print("Top 5 Female Percentages:")
print(top_5_female)
```

```python
# Save the results to new CSV files
top_5_biblical_male.to_csv('Baby_Names_Start/top_5_male_biblical_name_percentages.csv', index=False)
top_5_biblical_female.to_csv('Baby_Names_Start/top_5_female_biblical_name_percentages.csv', index=False)
```

```python
import matplotlib.pyplot as plt
# Create pie charts for the top 5 male and top 5 female percentages

# Top 5 male percentages
top_5_male_labels = top_5_male['Name']
top_5_male_sizes = top_5_male['Percentage']

# Top 5 female percentages
top_5_female_labels = top_5_female['Name']
top_5_female_sizes = top_5_female['Percentage']

# Create pie chart for male names
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.pie(top_5_male_sizes, labels=top_5_male_labels, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Male Biblical Names')

# Create pie chart for female names
plt.subplot(1, 2, 2)
plt.pie(top_5_female_sizes, labels=top_5_female_labels, autopct='%1.1f%%', startangle=140)
plt.title('Top Female Biblical Names')

# Show the pie charts
plt.tight_layout()
plt.show()
```
