```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Fetch the webpage content
url = 'https://www.ssa.gov/oact/babynames/decades/names1920s.html'
response = requests.get(url)
webpage = response.content

# Step 2: Parse the HTML
soup = BeautifulSoup(webpage, 'html.parser')

# Step 3: Extract the table data
table = soup.find('table', {'class': 't-stripe'})
rows = table.find_all('tr')

data = []
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele])

# Remove empty rows (if any)
data = [row for row in data if row]

# Step 4: Convert to DataFrame
columns = ['Rank', 'Male Name', 'Number of Males', 'Female Name', 'Number of Females']
df = pd.DataFrame(data, columns=columns)

# Step 5: Save DataFrame to CSV
df.to_csv('C:/Users/Jeremiah/Homework/New folder/baby_names_1920s.csv', index=False)

# Step 6: Display the DataFrame
df.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the base URL and the range of decades
base_url = 'https://www.ssa.gov/oact/babynames/decades/names{}s.html'
decades = range(1910, 2020, 10)

# Loop through each decade, scrape the data, and save it to a CSV file
for decade in decades:
    url = base_url.format(decade)
    response = requests.get(url)
    webpage = response.content
    
    # Parse the HTML
    soup = BeautifulSoup(webpage, 'html.parser')
    
    # Extract the table data
    table = soup.find('table', {'class': 't-stripe'})
    rows = table.find_all('tr')
    
    data = []
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append([ele for ele in cols if ele])
    
    # Remove empty rows (if any)
    data = [row for row in data if row]
    
    # Convert to DataFrame
    columns = ['Rank', 'Male Name', 'Number of Males', 'Female Name', 'Number of Females']
    df = pd.DataFrame(data, columns=columns)
    
    # Save DataFrame to CSV
    filename = f'C:/Users/Jeremiah/Homework/New folder/baby_names_{decade}s.csv'
    df.to_csv(filename, index=False)
    
    # Display a message indicating the progress
    print(f'Saved baby names for the {decade}s to {filename}')

# Display the first few rows of the last DataFrame as a check
df.head()
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# URL for the baby names data
url = "https://www.ssa.gov/oact/babynames/top5names.html"

# Fetch the HTML content
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Extract the table data
table = soup.find("table")
rows = table.find_all("tr")[1:]  # Skip the header row

# Initialize lists to store data
years = []
female_names = []
male_names = []

# Parse the table rows
for row in rows:
    cols = row.find_all("td")
    if len(cols) >= 6:
        year = cols[0].text.strip()
        female_name1 = cols[1].text.strip()
        female_name2 = cols[2].text.strip()
        female_name3 = cols[3].text.strip()
        female_name4 = cols[4].text.strip()
        female_name5 = cols[5].text.strip()
        male_name1 = cols[6].text.strip()
        male_name2 = cols[7].text.strip()
        male_name3 = cols[8].text.strip()
        male_name4 = cols[9].text.strip()
        male_name5 = cols[10].text.strip()

        years.append(year)
        female_names.append(f"{female_name1}, {female_name2}, {female_name3}, {female_name4}, {female_name5}")
        male_names.append(f"{male_name1}, {male_name2}, {male_name3}, {male_name4}, {male_name5}")

# Create a DataFrame
df = pd.DataFrame({
    "Year": years,
    "Female Names": female_names,
    "Male Names": male_names
})

# Write to CSV
df.to_csv("C:/Users/Jeremiah/Homework/New folder/2top_baby_names.csv", index=False)
print("Data saved to top_baby_names.csv")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# URL for the baby names data
url = "https://www.ssa.gov/oact/babynames/top5names.html"

# Fetch the HTML content
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Extract the table data
table = soup.find("table")
rows = table.find_all("tr")[1:]  # Skip the header row

# Initialize lists to store data
years = []
female_names = []
male_names = []
ranks = []  # New list for ranks

# Parse the table rows
for row in rows:
    cols = row.find_all("td")
    if len(cols) >= 11:
        year = cols[0].text.strip()
        female_name1 = cols[1].text.strip()
        female_name2 = cols[2].text.strip()
        female_name3 = cols[3].text.strip()
        female_name4 = cols[4].text.strip()
        female_name5 = cols[5].text.strip()
        male_name1 = cols[6].text.strip()
        male_name2 = cols[7].text.strip()
        male_name3 = cols[8].text.strip()
        male_name4 = cols[9].text.strip()
        male_nmae5 = cols[10].text.strip()

        years.append(year)
        female_names.append(f"{female_name1}, {female_name2}, {female_name3}, {female_name4}, {female_name5}")
        male_names.append(f"{male_name1}, {male_name2}, {male_name3}, {male_name4}, {male_name5}")

        # Add ranks (1 to 5) to the list
        ranks.extend([1, 2, 3, 4, 5])

# Create a DataFrame
df = pd.DataFrame({
    "Year": years,
    "Rank": ranks,  # New column for ranks
    "Female Names": female_names,
    "Male Names": male_names
})

# Write to CSV
df.to_csv("top_baby_names_with_ranks.csv", index=False)
print("Data saved to top_baby_names_with_ranks.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'https://www.behindthename.com/names/usage/all-biblical'

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')
names = soup.find_all('span', class_='listname')

with open('biblical_names.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Name'])  # Write the header row
    for name in names:
        writer.writerow([name.text])

print("Names have been successfully scraped and saved to 'biblical_names.csv'")
```
