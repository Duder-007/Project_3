```python
# Scrapping the most important events table from "https://news.gallup.com/poll/3427/most-important-events-century-from-viewpoint-people.aspx"
# importing packages

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://news.gallup.com/poll/3427/most-important-events-century-from-viewpoint-people.aspx"
response = requests.get(url)
webpage = response.content

soup = BeautifulSoup(webpage, 'html.parser')
```

```python
rows = soup.table.select("tr")
data_rows = rows[2:]

data = []
for row in data_rows:
    row_data = (row.stripped_strings)
    data.append(row_data)

Events_df = pd.DataFrame(data)
```

```python
Events_df.rename(columns={0 : "Event", 
                          1 : "% Viewed Most Important", 
                          2 : "% Viewed Important", 
                          3 : "% Viewed Somewhat Important", 
                          4 : "% Viewed Not Important", 
                          5 : "% Dont Know"}, 
                          inplace = True)
```

```python
# Replace the * with >0.5, value come from gallup table legend
Events_df["% Dont Know"] = Events_df["% Dont Know"].replace("*", ">0.5")
```
